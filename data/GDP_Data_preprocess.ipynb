```python

# Read the CSV file into a DataFrame
df = pd.read_csv('HDI_trends.csv', encoding="ISO-8859-1")

# Define the columns to be melted (in this case, columns 2 to 9)
value_vars = df.columns[2:10]

# Use the pd.melt() function to reshape the DataFrame
df_long = pd.melt(df, id_vars=['HDI rank', 'Country'], value_vars=value_vars, var_name='year', value_name='HDI')

# Sort the DataFrame by 'HDI rank' and 'year'
df_long.sort_values(by=['HDI rank', 'year'], inplace=True)

# Reset the index
df_long.reset_index(drop=True, inplace=True)

df_long.rename(columns={'year': 'Year'}, inplace=True)

# Display the resulting DataFrame
print(df_long)
```

```python
df_long.to_csv("HDI_trend_data.csv", index=False)
```
